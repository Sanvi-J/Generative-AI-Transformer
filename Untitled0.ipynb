```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

class AND(object):
    def __init__(self,x,y):
        self.x =  np.array(x)
        self.y = np.array(y)
        z = self.x.shape
        self.w = np.random.rand(1, z[1]) * 0.01
        self.bias = np.random.rand(1) * 0.01



    def sigmoid(self,a):
        return 1 / (1 + np.exp(-a))

    def sigmoid_derivative(self, a):
        return self.sigmoid(a) * (1 - self.sigmoid(a))

    def cost_calculation(self, y_hat, y):
        m = len(y)
        cost = (1 / (2 * m)) * (np.sum((y_hat - y) ** 2))
        return cost

    def feedforward(self, x):
        self.z = np.dot(self.w,x) + self.bias
        self.yHat = self.sigmoid(self.z)

        return self.yHat, self.z


    def train(self, epochs=1000):
        samples=self.x.shape[0]
        costs=[]
        for epoch in range(epochs):
            total_cost=0
            for i in range(len(x)):
                inputs = self.x[i].T
                output = self.y[i]
                yhat,z= self.feedforward(inputs)
                total_cost+= self.cost_calculation(yhat,output)

                dJ_dz = (yhat - output) * self.sigmoid_derivative(z)
                dJ_dw =  dJ_dz*inputs.T

                alpha = 0.1

                self.w -= alpha * dJ_dw
                self.bias -= alpha * dJ_dz.item()

            costs.append(total_cost/samples)

        plt.plot(np.arange(epochs), costs)
        plt.ylabel("Cost")
        plt.xlabel("Epoch")

```

```python

class XOR(object):
    def __init__(self,x,hidden,y):
        self.x =  np.array(x)
        self.y = np.array(y)
        input_size=self.x.shape[1]
        output_size=self.y.shape[1]
        self.w1 = np.random.rand(hidden,input_size) * 0.01
        self.b1 = np.random.rand(hidden,1) * 0.01
        self.w2 = np.random.rand(output_size, hidden) * 0.01
        self.b2 = np.random.rand(output_size,1) * 0.01


    def sigmoid(self,a):
        return 1 / (1 + np.exp(-a))

    def sigmoid_derivative(self, a):
        return self.sigmoid(a) * (1 - self.sigmoid(a))

    def cost_calculation(self, y_hat, y):
        m = len(y)
        cost = (1 / (2 * m)) * (np.sum((y_hat - y) ** 2))
        return cost

    def feedforward(self, x):
        self.z1 = np.dot(self.w1,x) + self.b1
        self.yHat1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.w2,self.yHat1) + self.b2
        self.yHat2 = self.sigmoid(self.z2)

        return self.yHat1, self.z1, self.yHat2, self.z2


    def train(self, epochs=1000):
        samples=self.x.shape[0]
        costs=[]
        for epoch in range(epochs):
            total_cost=0
            for i in range(len(self.x)):
                inputs = self.x[i].reshape(-1,1)
                outputs = self.y[i].reshape(-1,1)
                yhat1,z1,yhat2,z2= self.feedforward(inputs)
                total_cost+= self.cost_calculation(yhat2,outputs)

                dJ_dz_2 = (yhat2 - outputs) * self.sigmoid_derivative(z2)
                dJ_dw_2 = np.dot(dJ_dz_2, yhat1.T)

                dJ_dz_1 = np.dot(self.w2.T, dJ_dz_2) * self.sigmoid_derivative(z1)
                dJ_dw_1 = np.dot(dJ_dz_1, inputs.T)


                alpha = 0.1
                self.w2 -= alpha * dJ_dw_2
                self.b2 -= alpha * dJ_dz_2
                self.w1 -= alpha * dJ_dw_1
                self.b1 -= alpha * dJ_dz_1.reshape(self.b1.shape)

            costs.append(total_cost/samples)

        plt.plot(np.arange(epochs), costs)
        plt.ylabel("Cost")
        plt.xlabel("Epoch")
IL = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
OL = np.array([[0], [1], [1], [0]])
X = XOR(IL, 2, OL)
X.train()

```
